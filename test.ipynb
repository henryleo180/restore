```python
def add(df):
    previous_names = set()
    result = []

    # Iterate through the dataset
    for index, row in df.iterrows():
        name = row['IDRSSD']
        date = row['Reporting_Period_End_Date']

        if name not in previous_names:
            result.append({'name': name, 'come in': 4 * (date.year - 2018) + (date.month - 3) / 3, 'date': date})
            previous_names.add(name)

        

    for entry in result:
        print(f"Name: {entry['name']}, Event: {entry['come in']}, Date: {entry['date']}")

    result_df = pd.DataFrame(result)

    #result_df = result_df[result_df['date'] != '2018-03-31']

    return result_df

#disappeared_names = disapp(bank_data)
result = add(bank_data)

grouped_data = bank_data.groupby('Reporting_Period_End_Date')['Financial_Institution_Name'].nunique().reset_index(name='Count')

print(grouped_data)

#df2 = pd.DataFrame(disappeared_names.items(), columns=['Financial_Institution_Name', 'Reporting_Period_End_Date'])

#df2['Score'] = 1
```

```python
# Lasso
## feature selection
## simple reg using traditional split
#$$$ TBC: time series split

from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

X = final.iloc[:, 2:-1]
y = final.iloc[:, -1]
ids = final.iloc[:, 0]
times = final.iloc[:, 1]

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test, ids_train, ids_test, times_train, times_test = train_test_split(X, y, ids, times, test_size=0.2, random_state=42)

lasso = LassoCV(cv=5, random_state=0, max_iter=900000).fit(X_train, y_train)

selected_features = final.columns[2:-1][lasso.coef_ != 0]

X_train_selected = X_train[:, lasso.coef_ != 0]
X_test_selected = X_test[:, lasso.coef_ != 0]

y_pred = lasso.predict(X_test)

results = pd.DataFrame({'ID': ids_test, 'time':times_test,'y_pred': y_pred})

print(results)


results.to_csv('results.csv', index=False)

results[results['y_pred'] > 0.03].to_csv('results2.csv', index=False)
```
